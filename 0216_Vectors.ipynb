```
def plot_vectors(vecs):
  colors =['r','b','g','y']
  i=0
  for vec in vecs:
    plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=1,color=colors[i%len(colors)])
    i += 1
    print(i)
  plt.xlim(-10,10)
  plt.ylim(-10,10)  
  plt.show()
```

```
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,-3,-2),(0,0,4,-1)])
```

#Vector Dot Product

```
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-3,-2]),np.asarray([0,0,4,-1])]
```

```
plot_vectors(vecs)
```

```
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

$\vec{a}\cdot\vec{b}=|\vec{a}| |\vec{b}| \cos(\theta)=a_x
b_x + a_y b_y$

```
a_dot_b = np.dot(a,b)
```

```
print(a_dot_b)
```

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}| |\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|}$

```
a_b = np.dot(a,b)/np.linalg.norm(b)
```

```
print(a_b)
```

Unit vector along direction b

$\vec{a_b} = a_b \hat{b} = a_b \frac{\vec{b}}{|\vec{b}|}$

```
vec_a_b = (a_b/np.linalg.norm(b))*b
print(vec_a_b)
```

###The dot product $\vec{a}.\vec{b}$ may be positive or negative or zero

####i. If the angle between the two vectors is acute (i.e., 0<$\theta$<$90^\circ$) then cos$\theta$ is positive. In this case dot product is positive.

####ii.  If the angle between the two vectors is obtuse (i.e., 90<$\theta$<$180^\circ$) then cos$\theta$ is negative. In this case dot product is negative.

####iii.  If the angle between the two vectors is $90^\circ$ (i.e., $\theta$=$90^\circ$) then cos$\theta$=cos$90^\circ$=0. In this case dot product is zero.
